```python
import os
import pandas as pd
from natsort import natsorted
import glob
```

```python
# Función para procesar un archivo .txt y obtener las dos columnas
def procesar_archivo_txt(archivo):
    with open(archivo, 'r') as f:
        lines = f.readlines()

    # Filtrar las líneas que comienzan con números
    data_lines = [line.strip().split() for line in lines if line[0].isdigit()]

    X = [x[0] for x in data_lines]
    longitud_onda = [y[1] for y in data_lines]

    # Crear un DataFrame con las dos columnas
    df = pd.DataFrame([longitud_onda], columns=X)
    return df

# Ruta del directorio
directorio = "/content/"
patron_txt = "*.txt"

dicc={}
# Lista de archivos .txt en el directorio
archivos_txt = natsorted(glob.glob(directorio + patron_txt))
for i, j in enumerate(archivos_txt):
  dicc[i] = procesar_archivo_txt(j)

# Lista de archivos .txt en el directorio de manera individual
archivos_txt = natsorted(Path(directorio).glob(patron_txt))
# Obtener solo los nombres de los archivos (sin extensión)
nombres_archivos = [archivo.stem for archivo in archivos_txt]

# Concatenar los DataFrames por filas desde el diccionario
df_resultado = pd.concat(dicc.values(), ignore_index=True)

df_excel = df_resultado.set_index(pd.Index(nombres_archivos))
df_excel
```
